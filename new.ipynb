```python
import pandas as pd
import numpy as np


depth_df['Time'] = depth_df['Time'].str.replace(r'\s\+\d+\s\w+', '', regex=True)  # Remove timezone and text
depth_df['Time'] = pd.to_datetime(depth_df['Time'], errors='coerce')  # Convert to datetime

depth_df.dropna(subset=['Time'], inplace=True)

# Define the function to add price features
def add_price_features(df):
    df = df.copy()
    df = df.sort_values('Time')  

    
    df['mid_price'] = (df['BidPriceL1'] + df['AskPriceL1']) / 2

    
    df['log_return'] = np.log(df['mid_price'] / df['mid_price'].shift(1))

    
    df.set_index('Time', inplace=True)
    df['volatility_10s'] = df['log_return'].rolling('10s').std()
    df['volatility_30s'] = df['log_return'].rolling('30s').std()
    df.reset_index(inplace=True)

    return df


df1= add_price_features(depth_df)


print(df1[['Time', 'mid_price', 'log_return', 'volatility_10s', 'volatility_30s']].head(10))
```

```python
def add_volume_features(trades):
    trades = trades.copy()
    trades.set_index('Time', inplace=True)

   
    trades['buy_volume'] = trades['Quantity'].where(trades['IsMarketMaker'] == False, 0)
    trades['sell_volume'] = trades['Quantity'].where(trades['IsMarketMaker'] == True, 0)

    
    print("Buy Volume Sample:")
    print(trades[['buy_volume']].head())
    print("Sell Volume Sample:")
    print(trades[['sell_volume']].head())

    
    resampled = trades.resample('1s').agg({
        'buy_volume': 'sum',
        'sell_volume': 'sum',
        'Price': ['mean', 'sum'],
        'Quantity': 'sum'
    })

    
    resampled.columns = ['_'.join(col).strip() for col in resampled.columns.values]

    
    print("Resampled DataFrame Sample:")
    print(resampled.head())

    
    resampled['volume_imbalance'] = resampled['buy_volume_sum'] - resampled['sell_volume_sum']

    
    resampled['vwap'] = resampled['Price_sum'] / resampled['Quantity_sum']

    
    resampled['vwap_change_10s'] = resampled['vwap'].diff(10)
    resampled['vwap_change_30s'] = resampled['vwap'].diff(30)

    resampled.reset_index(inplace=True)
    return resampled
```

```python
from sklearn.preprocessing import StandardScaler


features_only = clean_merged_df.drop(columns=['Time_x', 'Time_y', 'Time_rounded'])

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features_only)


scaled_df = pd.DataFrame(scaled_features, columns=features_only.columns)
```

```python
from sklearn.decomposition import PCA
import numpy as np

# we are using iterative method to find the optimal number of components
pca = PCA()
pca.fit(scaled_df)


cumulative_variance = np.cumsum(pca.explained_variance_ratio_)


threshold = 0.95  
optimal_components = np.argmax(cumulative_variance >= threshold) + 1

print(f"Optimal number of components to retain {threshold*100}% variance: {optimal_components}")
```

```python
pca = PCA(n_components=optimal_components)
pca_result = pca.fit_transform(scaled_df)


pca_df = pd.DataFrame(pca_result, columns=[f'PC{i+1}' for i in range(optimal_components)])
```

```python
import matplotlib.pyplot as plt

if optimal_components >= 2:
    plt.figure(figsize=(8, 6))
    plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.5)
    plt.title('PCA-Reduced Data (First 2 Components)')
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.grid(True)
    plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


range_n_clusters = list(range(2, 10))
scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(pca_result)
    score = silhouette_score(pca_result, labels)
    scores.append(score)


plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, scores, marker='o')
plt.title("Silhouette Score vs Number of Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()


best_k = range_n_clusters[np.argmax(scores)]
print(f"Best number of clusters: {best_k}")
```

```python
import hdbscan
import numpy as np
import pandas as pd
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import product
from tqdm import tqdm


X = pca_df.copy()


min_cluster_sizes = [20, 30, 40, 50]
min_samples_list = [5, 10, 15, 20]


best_score = -1
best_params = {}
best_labels = None

results = []
# we are using iterative method to find optimal values of parameters.
print("Running HDBSCAN grid search")
for min_cluster_size, min_samples in tqdm(product(min_cluster_sizes, min_samples_list)):
    try:
        clusterer = hdbscan.HDBSCAN(min_cluster_size=min_cluster_size, min_samples=min_samples)
        labels = clusterer.fit_predict(X)

        
        mask = labels != -1
        if mask.sum() > 1 and len(set(labels[mask])) > 1:
            score = silhouette_score(X[mask], labels[mask])
            results.append((min_cluster_size, min_samples, score))

            if score > best_score:
                best_score = score
                best_params = {'min_cluster_size': min_cluster_size, 'min_samples': min_samples}
                best_labels = labels
    except Exception as e:
        print(f"Error with min_cluster_size={min_cluster_size}, min_samples={min_samples}: {e}")


print("\nâœ… Best HDBSCAN parameters:")
print(f"min_cluster_size: {best_params['min_cluster_size']}, min_samples: {best_params['min_samples']}")
print(f"Silhouette score: {best_score:.4f}")


score_df = pd.DataFrame(results, columns=['min_cluster_size', 'min_samples', 'silhouette_score'])

plt.figure(figsize=(10, 6))
pivot_table = score_df.pivot("min_cluster_size", "min_samples", "silhouette_score")
sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("HDBSCAN Silhouette Scores")
plt.show()

# Final model using best parameters
final_clusterer = hdbscan.HDBSCAN(**best_params)
final_labels = final_clusterer.fit_predict(X)

# Count of each cluster
unique, counts = np.unique(final_labels, return_counts=True)
print("\nðŸ“Š Cluster Counts (including noise as -1):")
for u, c in zip(unique, counts):
    print(f"Cluster {u}: {c} points")

# Visualize clusters 
if X.shape[1] >= 2:
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=final_labels, cmap='tab10', s=20, alpha=0.7)
    plt.colorbar(scatter, label="Cluster Label")
    plt.title("HDBSCAN Clustering Result (PCA-Reduced)")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid(True)
    plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


range_n_clusters = list(range(2, 10))
gmm_scores = []
db_scores = []
gmm_models = []

for n in range_n_clusters:
    gmm = GaussianMixture(n_components=n, covariance_type='full', random_state=42)
    labels = gmm.fit_predict(pca_df)
    
    sil_score = silhouette_score(pca_df, labels)
    db_index = davies_bouldin_score(pca_df, labels)
    
    gmm_scores.append(sil_score)
    db_scores.append(db_index)
    gmm_models.append((gmm, labels))

#  Plotting Silhouette Scores
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range_n_clusters, gmm_scores, marker='o', color='green')
plt.title("Silhouette Score (GMM)")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)

#  Plotting Davies-Bouldin Index
plt.subplot(1, 2, 2)
plt.plot(range_n_clusters, db_scores, marker='o', color='orange')
plt.title("Davies-Bouldin Index (GMM)")
plt.xlabel("Number of Clusters")
plt.ylabel("DB Index (Lower = Better)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Auto-Suggest Best Cluster Count
best_n_silhouette = range_n_clusters[np.argmax(gmm_scores)]
best_n_db = range_n_clusters[np.argmin(db_scores)]

print(f"ðŸ” Best by Silhouette Score: {best_n_silhouette} (Score: {max(gmm_scores):.4f})")
print(f"ðŸ” Best by Davies-Bouldin Index: {best_n_db} (Index: {min(db_scores):.4f})")
```

```python
# adding some features 
clean_merged_df['spread'] = clean_merged_df['AskPriceL1'] - clean_merged_df['BidPriceL1']
clean_merged_df['mid_price'] = (clean_merged_df['BidPriceL1'] + clean_merged_df['AskPriceL1']) / 2
clean_merged_df['log_return'] = np.log(clean_merged_df['mid_price'] / clean_merged_df['mid_price'].shift(1))
clean_merged_df['volatility_10s'] = clean_merged_df['log_return'].rolling(window=10, min_periods=1).std()

# Group by regime and analyze
regime_stats = clean_merged_df.groupby('regime').agg({
    'volatility_10s': 'mean',
    'spread': 'mean',
    'mid_price': ['first', 'last'],
    'Time_x': ['count', 'min', 'max']
})

# Calculate price direction (trend) = sign of price change
regime_stats[('mid_price', 'trend')] = np.sign(regime_stats[('mid_price', 'last')] - regime_stats[('mid_price', 'first')])
regime_stats.columns = ['_'.join(col).strip() for col in regime_stats.columns.values]
regime_stats.reset_index(inplace=True)

print(regime_stats)
```

## Strategy Ideas Based on Regimes

1. **Regime 0** - Mean Reverting & Illiquid:  
   Consider range-bound strategies like Bollinger Bands or pairs trading.

2. **Regime 1** - Volatile & Trendy:  
   Momentum strategies (e.g., moving average crossover, breakout trades).

3. **Regime 2** - Stable & Liquid:  
   Consider market-making or arbitrage due to narrow spread and stability.
